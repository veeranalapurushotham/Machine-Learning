```python
import numpy as np
import pandas as pd 
titanic_data=pd.read_csv('titanic.csv')
titanic_data.head()
```

**checking null values in the dataset**

```python
titanic_data.isnull().sum()
```

```python
titanic_data.shape
```

**here cabin and age has most null values becuase that i remove cabin feature and replace null with other values in age feature**

```python
titanic_data.head()
```

```python

titanic_data.drop(['Name','Ticket','Cabin'], axis=1, inplace=True, errors='ignore')
titanic_data['Age'].interpolate(method='linear', inplace=True)
titanic_data.head()
```

*** ml models can only understand the numerical values ,so i sex features it contains string to num**

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
titanic_data['Sex']=lb.fit_transform(titanic_data['Sex'])
titanic_data.head()
```

```python
titanic_data['Embarked'].isnull().sum()
```

```python
titanic_data.dropna(subset=['Embarked'], inplace=True)
```

```python
titanic_data.isnull().sum()
```

**Now embarked is convert numerical values based on the category**

```python
dummies = pd.get_dummies(titanic_data['Embarked'], prefix='Embarked', drop_first=True)
titanic_data = pd.concat([titanic_data, dummies], axis=1)
titanic_data.drop('Embarked', axis=1, inplace=True)
titanic_data.head()
```

**now convert embarked boolean values into numeric**

```python
titanic_data['Embarked_Q']=titanic_data['Embarked_Q'].astype(int)
titanic_data['Embarked_S']=titanic_data['Embarked_S'].astype('category')
titanic_data['Embarked_S']=titanic_data['Embarked_S'].cat.codes
titanic_data.head()
```

```python
x=titanic_data.drop('Survived', axis=1)
y=titanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.head()
```

```python
y_train.head()
```

<spam style="color:magneta">Decision Tree Model</spam>

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
x_test.head()
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
model.predict([[282,3,1,28.0,0,0,7.8545,0,1]])
```

```python
from sklearn.metrics import accuracy_score
print("accuracy score for decision tree model:", accuracy_score(y_test, y_pred))
```
