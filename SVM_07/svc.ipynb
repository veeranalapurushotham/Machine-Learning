```python
import pandas as pd 
import numpy as np
from sklearn.datasets import load_digits
digits = load_digits()
dir(digits)
```

```python
digits.data
```

```python
digits.target
```

```python
digits.feature_names
```

```python
dataset=pd.DataFrame(digits.data,columns=digits.feature_names)
dataset['target']=digits.target
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
x=dataset.drop('target',axis=1)
y=dataset['target']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.svm import SVC
model=SVC(kernel='poly',C=10,gamma='scale')
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

***âšªnow accuracy of the model by predicting the x test data and comapring y test data***

```python
model.score(x_test,y_test)
```
