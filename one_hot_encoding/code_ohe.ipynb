```python
import numpy as np
import pandas as pd 
data=pd.read_csv('carprices.csv')
data
```

 ***here car model are nomial categorial values linear regression working with only numerial values.so now we need convert the categorial into numerical***


```python
dummy=pd.get_dummies(data['Car Model'],drop_first=True)
dummy
```

***now merging with dummy with data and remove the car model column becaus it's already in dummies,
and after we remove one of the dummy because to avoid dummy variable trap***

```python
connect=pd.concat([data,dummy],axis='columns')
final_data=connect.drop(['Car Model'],axis='columns')
final_data
```

 ***by doing manualy predictions***

```python
x=final_data.drop(['Sell Price($)'],axis='columns')
y=final_data['Sell Price($)']
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
print("the predicted price of mercedz benz with 4 year old mileageis 45000: ",model.predict([[45000,4,0,1]]))
print("the predicted price of bmw with 7 year old mileageis 86000: ",model.predict([[86000,7,1,0]]))
```

***now predict the model accuracy with predicted values with actual and return the score***

```python
model.score(x,y)
```

```python

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Car Model']=le.fit_transform(data['Car Model'])
data
```

```python
X=data[['Car Model','Mileage','Age(yrs)']]
Y=data['Sell Price($)']
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ohe=ColumnTransformer(transformers=[('cat',OneHotEncoder(),[0])],remainder='passthrough')
X=ohe.fit_transform(X)
X=X[:,1:]
oht_model=model.fit(X,Y)
X
```

```python
print("predict price of mercedz benz with 4 year old mileageis 45000: ",oht_model.predict([[0,1,45000,4]]))
print("the predicted price of bmw with 7 year old mileageis 86000: ",oht_model.predict([[0,0,86000,7]]))
```

```python
oht_model.score(X,Y)
```
